```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt

X,y = datasets.make_regression(n_samples=100, n_features=1, noise = 20, random_state=4)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2, random_state=1234)

fig = plt.figure(figsize=(8,6))
plt.scatter(X, y, color='b', marker='o', s=40)
plt.show()
```

```python
from linear_regression import LinearRegression, mse
regressor = LinearRegression(lr=0.1)

regressor.fit(X_train,y_train)
y_predicted = regressor.predict(X_test)
y_predicted

mse_val = mse(y_test, y_predicted)
mse_val
```

```python
y_pred_line = regressor.predict(X)
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8,6))
plt.scatter(X_train, y_train, cmap = 0.2, s=10)
plt.scatter(X_test, y_test, cmap = 0.1, s= 40)
plt.plot(X,y_pred_line, color = 'black', linewidth=2, label = 'prediction')
plt.show()
```
